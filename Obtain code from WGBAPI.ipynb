```python
#to obtain the inflation, consumer prices (anual %) on dataframe
inflation_df = wb.data.DataFrame('FP.CPI.TOTL.ZG')
```

```python
inflation_df
```

```python
inflation_df = wb.data.DataFrame('FP.CPI.TOTL.ZG', time = [2015,2016,2017,2018,2019,2020,2021])
inflation_df
```

```python
wb.region.info()
```

```python
#To obtain the values of inflation from years: 2015,2016,2017,2018,2019,2020,2021
#And this values are only from Latin America & Caribbean we put
inflation_df_LATAM = wb.data.DataFrame('FP.CPI.TOTL.ZG',
                                       wb.region.members('LCN'), #LCN = Latin America & Caribbean
                                       time = [2015,2016,2017,2018,2019,2020,2021],
                                       labels = True)
inflation_df_LATAM
```

```python
inflation_df_LATAM.info()
```

<h2>obtain unemployment data:</h2>

```python
wb.series.info(q='unemployment')
```

```python
#Obtain Unemployment, total (% of total labor force) (national estimate)
up_df_LATAM = wb.data.DataFrame('SL.UEM.TOTL.NE.ZS', wb.region.members('LCN'), time = [2015,2016,2017,2018,2019,2020,2021])
up_df_LATAM
```

```python
wb.series.info(q='population')
```

```python
#Obtain Population growth (annual %)
popgrouth_df_LATAM = wb.data.DataFrame('SP.POP.GROW', wb.region.members('LCN'),#
                                       time = [2015,2016,2017,2018,2019,2020,2021], #años que voy a analizar
                                      labels = True ) #Labels=True muestra el nombre de los países
popgrouth_df_LATAM
```

```python
wb.series.info(q='Debt')
```

```python
#Obtain Central government debt, total (% of GDP)
totaldebt_df_LATAM = wb.data.DataFrame('GC.DOD.TOTL.GD.ZS', wb.region.members('LCN'),#
                                       time = [2015,2016,2017,2018,2019,2020,2021], #años que voy a analizar
                                      labels = True ) #Labels=True muestra el nombre de los países
totaldebt_df_LATAM
```
