```python
import wbgapi as wb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#To obtain the values of inflation from years: 2015,2016,2017,2018,2019,2020,2021
#And this values are only from Latin America & Caribbean we put
inflation_df_LATAM = wb.data.DataFrame('FP.CPI.TOTL.ZG',
                                       wb.region.members('LCN'), #LCN = Latin America & Caribbean
                                       time = [2015,2016,2017,2018,2019,2020,2021],
                                       labels = True)
```

```python
#We prepare the data to stay with the countries I want to study
inflation_df_LATAM.set_index('Country', inplace=True)
inflation_df_LATAM.sort_values('Country')
```

```python
#To get the inflation values of the years: 2015,2016,2017,2018,2019,2020,2021
#We like obtain data about: Argentina, Bolivia, Brazil, Chile, Colombia, Costa Rica,Dominican Republic, 
#                           Ecuador, El Salvador, Guatemala, Honduras, Mexico, Nicaragua, Panama,
#                           Paraguay, Peru, Uruguay
inf_data = inflation_df_LATAM.loc[['Argentina','Bolivia','Brazil','Chile','Colombia','Costa Rica',
                                   'Dominican Republic','Ecuador','El Salvador','Guatemala','Honduras',
                                   'Mexico', 'Nicaragua','Panama','Paraguay','Peru','Uruguay']]
inf_data
```

```python
#los datos están separados por ';' y no por ',' por lo que hay que acalarar que se separan así
pd.read_csv('inflation_ARG.csv', sep=';')
```

```python
#Voy a cambiar los valores a mano nomás, al ser pocos es más fácil
inf_data.loc['Argentina'][0]=23.9
inf_data.loc['Argentina'][1]=40.2
inf_data.loc['Argentina'][2]=24.8
inf_data.loc['Argentina'][3]=47.6
inf_data.loc['Argentina'][4]=53.8
inf_data.loc['Argentina'][5]=36.1
inf_data.loc['Argentina'][6]=50.9
inf_data
```

```python
#plot a line chart
Years=['YR2015','YR2016','YR2017','YR2018','YR2019','YR2020','YR2021']
Argentina= inf_data.loc['Argentina']
Uruguay= inf_data.loc['Uruguay']
Chile = inf_data.loc['Chile']
Paraguay = inf_data.loc['Paraguay']
Bolivia = inf_data.loc['Bolivia']
Brazil = inf_data.loc['Brazil']

#,'Chile','Colombia','Costa Rica',
#                                   'Dominican Republic','Ecuador','El Salvador','Guatemala','Honduras',
#                                   'Mexico', 'Nicaragua','Panama','Paraguay','Peru','Uruguay']]

plt.plot(Years,Argentina, label='Argentina')
plt.plot(Years,Uruguay, label='Uruguay')
plt.plot(Years,Chile, label='Chile')
plt.plot(Years,Paraguay, label='Paraguay')
plt.title('inflation per country')
plt.xlabel('Years')
plt.ylabel('inflation')
plt.legend()
plt.show()
```

```python
#plot a bar chart
Y=[['Argentina','Uruguay','Chile','Paraguay','Bolivia','Brazil']]
X=['YR2015','YR2016','YR2017','YR2018','YR2019','YR2020','YR2021']
plt.barh(Y,X)
plt.title('title name')
plt.ylabel('y axis name')
plt.xlabel('x axis name')
plt.show()
```
